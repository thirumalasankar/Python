Matplotlib
--------------


Matplotlib is a "python library" used to create 2D graphs and plots by using python scripts. It has a "module" named "pyplot" which makes things easy for plotting by providing feature to control line styles, font properties, formatting axes etc.

```python
import matplotlib.pyplot as plt
```

```python
days=list(range(1,8))
lappySales=[123,356,234,700,432,762,289]
```

```python
plt.title("BigBillionSales")
plt.xlabel("No of days")
plt.ylabel("No of Sales")
plt.plot(days,lappySales)
plt.show()
```

```python
plt.title("BigBillionSales",fontsize="20",color="blue")
plt.xlabel("No of days",fontsize="12")
plt.ylabel("No of Sales",fontsize="12")
plt.plot(days,lappySales,color="green",marker="o",markerfacecolor="black",linestyle="dashed")
plt.grid(axis='y')
plt.show()
```

```python
#subplot

import matplotlib.pyplot as plt
import numpy as np

#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 1, 1)
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 1, 2)
plt.plot(x,y)

plt.show()
```

```python
days=range(1,8)
lappySales=[123,455,762,982,178,468,189]
mobileSales=[589,142,475,680,277,272,282]
chocSales=[132,578,256,688,292,588,289]
plt.title("BigBillionSales",fontsize="20",color="blue")
plt.xlabel("No of days",fontsize="12")
plt.ylabel("No of Sales",fontsize="12")
plt.plot(days,lappySales,color="green",marker="o",markerfacecolor="black",linestyle="dashed")
plt.plot(days,mobileSales,color="red",marker="*",markerfacecolor="black",linestyle="-")
plt.plot(days,chocSales,color="cyan",marker="o",markerfacecolor="black",linestyle="-.")
plt.grid(color="black",linewidth="2")
plt.legend(labels=('lappySales','phoneSales','ChocSals'),loc="upper left")
plt.xlim(2,6)
plt.ylim(200,700)
plt.show()
```

different graphs representation

                   ---> ScatterPlotting
                   ---> BarGraph(vertical,horizontal)
                   ---> histograms
                   ---> pieCharts

# ScatterPlotting

```python
plt.scatter(days,lappySales,color="green",label="lappysales")
plt.scatter(days,mobileSales,color="blue",label="phoneSales")
plt.scatter(days,chocSales,color="black",label="chocSales")
plt.grid()
plt.legend(loc="upper right")
plt.show()
```

# Bar graph

```python
plt.bar(days,lappySales,color="green",width=.4,label="lappysales")
plt.legend(loc="upper right")
plt.show()
```

```python
plt.ylabel("no of days")
plt.xlabel("no of chocsales")
plt.barh(days,chocSales,color="blue",height=.2,label="chocsales")
plt.legend(loc="lower right")
plt.show()
```

# histograms

```python
plt.xlabel("no of days")
plt.ylabel("no of phonesales")
plt.hist(mobileSales,color="black",label="chocsales")
plt.show()
```

```python
#src -- gfg
# importing libraries
import matplotlib.pyplot as plt

# giving two age groups data
age_g1 = [1, 3, 5, 10, 15, 17, 18, 16, 19, 21,
		23, 28, 30, 31, 33, 38, 32, 40, 45,
		43, 49, 55, 53, 63, 66, 85, 80, 57,
		75, 93, 95]

age_g2 = [6, 4, 15, 17, 19, 21, 28, 23, 31, 36,
		39, 32, 50, 56, 59, 74, 79, 34, 98, 97,
		95, 67, 69, 92, 45, 55, 77, 76, 85]

# plotting first histogram
plt.hist(age_g1, label='Age group1', alpha=.7, color='red')

# plotting second histogram
plt.hist(age_g2, label="Age group2", alpha=.5,
		edgecolor='black', color='yellow')
plt.legend()

# Showing the plot using plt.show()
plt.show()
```

# pie chart

```python
celebrities = ["hannah","helmet","millie","kate","chris","shaun"]
votes = [20,20,20,20,20,20] 
plt.pie(votes,labels=celebrities,startangle=90)
plt.show()
```

```python
plt.pie(votes,labels=celebrities,autopct="%1.3f",radius=1.5,explode=[0.3,0.2,0,0,0,1],startangle=90,shadow=True)
plt.show()
```

```python
import numpy as np
x=np.linspace(-np.pi,np.pi,30)  #limits
cosWave=np.cos(x)
sinWave=np.sin(x)
ax=plt.axes([.1,0.1,.8,.8])    #axis defined
ax.plot(x,cosWave,label="cosWave",marker="*",linestyle="dashed")
ax.plot(x,sinWave,label="sinWave",marker="h")
plt.legend(loc="center right")
plt.show()
```
